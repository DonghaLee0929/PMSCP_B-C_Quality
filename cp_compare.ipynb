```python
max_time = 60 * 60
result = Datarecoder()
result.create_csv('very_high_large_family.csv')

for job_num in [10, 20, 30, 40, 50]:
    for machine_num in [2, 3, 4, 5]:
        for seed in range(5, 30, 5):
        # for seed in [5]:
            result.reset()

            print(f"\nJob: {job_num}, Machine: {machine_num}, Seed: {seed}")
        
            Gamma = 0.95
            Lambda = 1
            delta = 0.05

            environment = Environment(job_num, machine_num, math.floor(job_num/8)+1)
            env = environment.reset(seed=seed, tight='very_high', quality='very_high')

            best, _, Lambda, = framework(env, Gamma, Lambda, delta, recoder=result, assign_max_time=180, seq_max_time=180, max_time=max_time)
            print(f'Iteration num: {result.total_iter_num}, Elapsed time: {result.total_time}\n')
        
            result.cp_value, result.cp_time, result.cp_status, result.cp_feasible_sol, result.cp_feasible_time = global_cp(env, Gamma, Lambda, max_time, viz=True) # 수율과 위험도를 1000배 하여 정수화

            result.write_result_to_csv()
```
